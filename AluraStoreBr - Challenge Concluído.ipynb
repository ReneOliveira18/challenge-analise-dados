### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento


```python
#Calculando faturamento total de cada loja
faturamento_loja = loja['Preço'].sum()
faturamento_loja2 = loja2['Preço'].sum()
faturamento_loja3 = loja3['Preço'].sum()
faturamento_loja4 = loja4['Preço'].sum()

#printando os resultados
print(f'Faturamento da Loja: R$ {faturamento_loja:.2f}')
print(f'Faturamento da Loja 2: R$ {faturamento_loja2:.2f}')
print(f'Faturamento da Loja 3: R$ {faturamento_loja3:.2f}')
print(f'Faturamento da Loja 4: R$ {faturamento_loja4:.2f}')
```

# 2. Vendas por Categoria


```python
#analisando faturamento de cada categoria dentro de cada loja usando .groupby
#printando os resultados
#identificando categoria com maior faturamento dentro de cada loja .idxmax()
#retornando maior faturamento por loja .max()
#printando resultado por loja
#LOJA
analisar_loja = loja.groupby('Categoria do Produto')['Preço'].sum()
print(f'\nLoja 1\n',analisar_loja)
categoria_maior_faturamento_loja = analisar_loja.idxmax()
faturamento_max_loja = analisar_loja.max()
print(f'\n Loja1: Categoria com maior faturamento: {categoria_maior_faturamento_loja} (R${faturamento_max_loja:.2f})')


#LOJA2
analisar_loja2 = loja2.groupby('Categoria do Produto')['Preço'].sum()
print(f'\nLoja 2\n',analisar_loja2)
categoria_maior_faturamento_loja2 = analisar_loja2.idxmax()
faturamento_max_loja2 = analisar_loja2.max()
print(f'\n Loja2: Categoria com maior faturamento: {categoria_maior_faturamento_loja2} (R${faturamento_max_loja2:.2f})')

#LOJA3
analisar_loja3 = loja3.groupby('Categoria do Produto')['Preço'].sum()
print(f'\nLoja 3\n',analisar_loja3)
categoria_maior_faturamento_loja3 = analisar_loja3.idxmax()
faturamento_max_loja3 = analisar_loja3.max()
print(f'\n Loja3: Categoria com maior faturamento: {categoria_maior_faturamento_loja3} (R${faturamento_max_loja3:.2f})')

#LOJA4
analisar_loja4 = loja4.groupby('Categoria do Produto')['Preço'].sum()
print(f'\nLoja 4\n',analisar_loja4)
categoria_maior_faturamento_loja4 = analisar_loja4.idxmax()
faturamento_max_loja4 = analisar_loja4.max()
print(f'\n Loja4: Categoria com maior faturamento: {categoria_maior_faturamento_loja4} (R${faturamento_max_loja4:.2f})')
```

# 3. Média de Avaliação das Lojas

```python
media_avaliacao_loja = loja['Avaliação da compra'].mean()
media_avaliacao_loja2 = loja2['Avaliação da compra'].mean()
media_avaliacao_loja3 = loja3['Avaliação da compra'].mean()
media_avaliacao_loja4 = loja4['Avaliação da compra'].mean()

print(f'Média de Avaliação da Loja: {media_avaliacao_loja:.2f}')
print(f'Média de Avaliação da Loja 2: {media_avaliacao_loja2:.2f}')
print(f'Média de Avaliação da Loja 3: {media_avaliacao_loja3:.2f}')
print(f'Média de Avaliação da Loja 4: {media_avaliacao_loja4:.2f}')
```

# 4. Produtos Mais e Menos Vendidos

```python
#identificar e visualizar os produtos mais e menos vendidos por loja
def analisar_produtos_mais_menos_vendidos(df, loja, top_n=3):
  if 'Produto' not in df.columns:
    print(f"A coluna 'Produto' não foi encontrada no DataFrame da Loja {loja}.")
    return

#contando a quantidade de cada produto
  contagem_produtos = df['Produto'].value_counts().sort_values(ascending=False) #ascending=False ordena os valores do maior para o menor

#printando produto mais vendido
  print(f"\n -=-=- Top {top_n} Produto Mais Vendidos - {loja} -=-=-")
  if not contagem_produtos.empty:
    for produto, quantidade in contagem_produtos.head(top_n).items():
      print(f"- {produto} : {quantidade} vendas")
  else:
    print("Nenhum produto encontrado.")

#printando produto menos vendido
  print(f"\n -=-=- Top {top_n} Produto Menos Vendidos - {loja} -=-=-")
  if not contagem_produtos.empty:
    for produto, quantidade in contagem_produtos.tail(top_n).items():
      print(f"- {produto} : {quantidade} vendas")
  else:
    print("Nenhum produto encontrado.")

#chamando a função para cada loja
analisar_produtos_mais_menos_vendidos(loja,'Loja ')
analisar_produtos_mais_menos_vendidos(loja2,'Loja2 ')
analisar_produtos_mais_menos_vendidos(loja3,'Loja3 ')
analisar_produtos_mais_menos_vendidos(loja4,'Loja4 ')
```

# 5. Frete Médio por Loja


```python
#calculando custo médio do frete para cada loja
custo_medio_frete_loja = loja['Frete'].mean()
#printando os resultados
print(f"Custo Médio do Frete da Loja: R$ {custo_medio_frete_loja:.2f}")

#Loja2
custo_medio_frete_loja2 = loja2['Frete'].mean()
print(f"Custo Médio do Frete da Loja 2: R$ {custo_medio_frete_loja2:.2f}")

#Loja3
custo_medio_frete_loja3 = loja3['Frete'].mean()
print(f"Custo Médio do Frete da Loja 3: R$ {custo_medio_frete_loja3:.2f}")

#Loja4
custo_medio_frete_loja4 = loja4['Frete'].mean()
print(f"Custo Médio do Frete da Loja 4: R$ {custo_medio_frete_loja4:.2f}")
```
